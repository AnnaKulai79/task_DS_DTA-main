```python
np.random.seed(2)

n = 500
p = 0.10

A = stats.bernoulli.rvs(p, size=n)
B = stats.bernoulli.rvs(p, size=n)

# use stats.ttest_ind(B, A).pvalue,
result_ttest = stats.ttest_ind(B, A).pvalue
print(result_ttest) # статистично значущими є при pvalue <= 0,05
```

```python
import numpy as np
from scipy import stats


np.random.seed(42)

n = 500 

p_a = 0.10
p_b = 0.12

# use np.random.binomial to generate dist for p_a, p_b, with size n 
group_a = np.random.binomial(1, p_a, size=n)
group_b = np.random.binomial(1, p_b, size=n)

```

```python
# calculate mean, can use .mean()
group_a.mean(), group_b.mean()
```

```python
# create two variables and save result of stats.ttest_ind(<you group result B>, <you group result A>)
t_stat, p_value = stats.ttest_ind(group_b, group_a)
print(t_stat, p_value)
```
