**Task 4.** Двоє користувачів у просторі ознак:

> u1 = (перегляди, лайки) = (10, 2) \
> u2 = (15, 5)

Потрібно:
> знайти відстань між ними (евклідова), \
> зрозуміти, хто “активніший” у загальному масштабі.

```python
# 1. Відстань між користувачами
u1 = np.array([10, 2])
u2 = np.array([15, 5])
# use np.linalg.norm
# code 
print(np.linalg.norm(u1 - u2))

# 2. Рівень активності кожного (довжина вектора)
# use np.linalg.norm
# code 
print(np.linalg.norm(u1), np.linalg.norm(u2))
```

**Task 5.** Нормалізація поведінкових фіч перед кластеризацією


```python
import numpy as np

users = np.array([
    [10, 3, 1],
    [40, 2, 0],
    [20, 5, 2],
    [15, 1, 1],
    [60, 7, 3]
])

# use linalg.norm
# users / by previos step
norms = np.linalg.norm(users, axis=1, keepdims=True)
users_normalized = users / norms
print(users_normalized)
```

**Task 5.** Оцініть наскільки користувачу підходить товар

```python
# user data 
#u = (0.2, 0.8, 0.3, 0.1)

# product data
#p = (0.25, 0.75, 0.2, 0.15)


import numpy as np

u = np.array([0.2, 0.8, 0.3, 0.1])
m = np.array([0.25, 0.75, 0.2, 0.15])

score = np.dot(u,m) / (np.linalg.norm(u) * np.linalg.norm(m))# use np.dot 
print(score)
```
