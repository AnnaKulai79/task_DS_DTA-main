```python
# Task 0.3. Dot transpose matrix . Calculate diagonal
import numpy as np

# | user | sessions | watch_hours | spend |
# | ---- | -------- | ----------- | ----- |
# | 1    | 5        | 10          | 100   |
# | 2    | 10       | 20          | 200   |
# | 3    | 3        | 6           | 60    |
# | 4    | 8        | 16          | 160   |

X = np.array([
    [5, 10, 100],
    [10, 20, 200],
    [3, 6, 60],
    [8, 16, 160]
])

# use .T and @
print(X.T @ X)

# use by prev task np.diag
print(np.diag(X.T @ X))

# use np.linalg.matrix_rank 
print(np.linalg.matrix_rank(X))

```

```python
# Calculate weights
weights = np.array([0.5, 0.2, 0.01])
impact = X * weights
impact
```

```python
# L2-normalize
s = np.linalg.norm(X[:, 0])  # sessions
w = np.linalg.norm(X[:, 1])  # watch_hours
p = np.linalg.norm(X[:, 2])  # spend
print(s,w,p)
```
