Знайти median та mean. Що про них можна сказати?

```python
arr = [3, 3, 4, 100]
arr.sort() # sorted 
n = len(arr)
mean_ = sum(arr) / n
if  n % 2 == 1:
    median_ = arr[n // 2]
else:
    median_ = (arr[n // 2 - 1] + arr[n // 2]) / 2

print(mean_, median_)
```

**Задача 1** \
Дані: [10, 12, 12, 15, 20]\
Знайти моду.

```python
arr = [10, 12, 12, 15, 20]
dic = {}
for i in arr:
    if i in dic.keys():
        dic[i] += 1
    else:
        dic[i] = 1
max_ = max(dic.values())
lst_mode = []
for i in arr:
    if dic[i] == max_:
        lst_mode.append(i)
print(max_)
print(lst_mode)
```

Задача 2 

Кількість подій на користувача:
[1,1,1,2,3,100]

```python
from collections import Counter
arr = [1,1,1,2,2,2,3,100]
dic = Counter(arr)

arr.sort() # sorted 
n = len(arr)
mean_ = sum(arr) / n
if  n % 2 == 1:
    median_ = arr[n // 2]
else:
    median_ = (arr[n // 2 - 1] + arr[n // 2]) / 2
print(f"Mean = {mean_}, Median =  {median_}, Mode = {dic.most_common()}")
```

Задача 2 — два набори з однаковим середнім, але різним розкидом \
A: [4, 5, 6] \
B: [1, 5, 9] 

Середні однакові.  
Який набір стабільніший?

```python
a_set = [4, 5, 6]
b_set = [1, 5, 9]
def var_my(arr):
    n = len(arr)
    mean_ = sum(arr) / n
    return sum((i - mean_) ** 2 for i in arr) / n
print(var_my(a_set), var_my(b_set))
```

### Std 
Дані: [4, 5, 6, 9, 10]  

Порахуй variance і std. 


```python
arr = [4, 5, 6, 9, 10]
def var_my(arr):
    n = len(arr)
    mean_ = sum(arr) / n
    return sum((i - mean_) ** 2 for i in arr) / n
print(var_my(arr), var_my(arr) ** 0.5)
```
